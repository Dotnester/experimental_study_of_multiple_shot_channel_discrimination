```python
# Mitigation

from qiskit_ibm_runtime import QiskitRuntimeService
import datetime as dt
from typing import Dict, List

from UnifiedTester import TesterResultStorage

pre_mitigation_data = TesterResultStorage.load_from_directory("./IBM_BRISBANE_4_5")

service = QiskitRuntimeService()
backend = service.backend('ibm_brisbane')
calibration_date  = dt.datetime(2025, 5, 4)

properties = backend.properties(datetime=calibration_date)

from qiskit_aer.noise import NoiseModel

noise_model = NoiseModel.from_backend_properties(properties)

from qiskit_aer import AerSimulator

simulator = AerSimulator(noise_model=noise_model)

import mthree
mits: Dict[int, mthree.M3Mitigation] = dict()
for i in range(12):
    mits[i] = mthree.M3Mitigation(simulator)
    qubits = list(range(i+1))
    mits[i].cals_from_system(qubits)
```

```python
import re
pattern = r'Q(\d+)_L(\d+)_.*'
mitigated_quasi = list()

for i, name in enumerate(pre_mitigation_data.names):
    q_num, l_num = re.match(pattern, name).groups()
    mitigated_quasi.append(mits[int(q_num)-1].apply_correction(pre_mitigation_data.real_counts[i], list(range(int(q_num)))))

from collections import Counter
import numpy as np

total_shots = 10000

mitigated_counts_final:List[Dict[str, int]] = list()
for mit_quas in mitigated_quasi:
    bitstrings = list(mit_quas.keys())
    probs = np.array([mit_quas[bs] for bs in bitstrings])
    probs = np.clip(probs, 0, 1)  # remove any small negatives due to mitigation
    probs = probs / probs.sum()  # re-normalize to sum to 1

    sampled = np.random.choice(bitstrings, size=total_shots, p=probs)
    count = Counter(sampled)
    mitigated_counts_final.append(dict(count))
```

```python
# Plotting of mitigated results

import re
import matplotlib.pyplot as plt

total_shots = 20000 # 10k for rz and 10k for id
sum_right_answears = lambda circ_table: circ_table['TP'] + circ_table['TN'] + circ_table['GTP'] + circ_table['GTN']
pattern = r'Q(\d+)L(\d+)' # reg exp pattern for n of qubits and layers extraction

# process purely parallel circuits with short measurement
pure_parallel_processed_short = pure_parallel_short.process_results_short()
print("pure_parallel_processed_short")
print(pure_parallel_processed_short)
x_axis_vals1 = []
y_axis_vals1 = []
for name, item in pure_parallel_processed_short.items():
    q_num, l_num = re.match(pattern, name).groups()
    x_axis_vals1.append(int(q_num))
    y_axis_vals1.append(sum_right_answears(item)/total_shots)

# process purely parallel circuits with xor measurement
pure_parallel_processed_xor = pure_parallel_xor.process_results_xor()
print("pure_parallel_processed_xor")
print(pure_parallel_processed_xor)
x_axis_vals2 = []
y_axis_vals2 = []
for name, item in pure_parallel_processed_xor.items():
    q_num, l_num = re.match(pattern, name).groups()
    x_axis_vals2.append(int(q_num))
    y_axis_vals2.append(sum_right_answears(item)/total_shots)

plt.rcParams.update({
    "text.usetex": False,  # Disable LaTeX
    "font.family": "serif",
    "font.size": 14
})

plt.figure(figsize=(8, 6))

# Plot with thicker lines and updated colors
plt.plot(x_axis_vals1, y_axis_vals1, marker='o', linestyle='-', linewidth=2.5, color='#1f77b4', label=r'Short measurement')
plt.plot(x_axis_vals2, y_axis_vals2, marker='s', linestyle='--', linewidth=2.5, color='#d62728', label=r'XOR measurement')

# Labels and title
plt.xlabel(r'Number of copies', fontsize=20)
plt.ylabel(r'Probability', fontsize=20)
#plt.title(r'Correct gate identification by parallel scheme', fontsize=18)

# Ticks
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

# Grid and legend
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(loc="upper right", fontsize=16, frameon=True)

plt.tight_layout()
plt.savefig("./plots/pure_parallel_short_xor_mitigated.pdf", format="pdf", bbox_inches="tight")
plt.show()
```
